```python
import pandas as pd
```

```python
df=pd.read_csv("Heart.csv")
df
```

```python
df=df.drop(columns='Unnamed: 0')
```

```python
df
```

```python
df.info()
```

```python
df['ChestPain']=df['ChestPain'].astype('category')
df['ChestPain']=df['ChestPain'].cat.codes
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
#Convert category to numeric 
df['Thal']=df['Thal'].astype('category')
df['Thal']=df["Thal"].cat.codes
df
```

```python
df
```

```python
X=df.drop(columns='AHD')
X
```

```python
y=df['AHD']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=21)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
```

```python
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression(random_state=0).fit(X_train_scaled,y_train) #fit scled data
```

```python
log_reg.predict(X_train_scaled) #if we give X_train it predict y_train
```

```python
log_reg.score(X_train_scaled,y_train)
```

```python
log_reg.score(X_test_scaled,y_test)
```
